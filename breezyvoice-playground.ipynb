```python
import subprocess

# Run nvidia-smi command and get the output
try:
    result = subprocess.run(['nvidia-smi', '--query-gpu=name', '--format=csv,noheader'], stdout=subprocess.PIPE, text=True)
except:
    raise Exception("No GPU detected, please choose GPU as the accelerator: Settings > Accelerator") 
gpus = result.stdout.strip().split('\n')

# Assert that there are >= 1 GPUs
assert len(gpus) >= 1, f"Expected >=1 GPUs, but found {len(gpus)} please choose GPU as the accelerator: Settings > Accelerator"
print(f"Number of GPUs: {len(gpus)}")
```
